    inventario.append({
        "nombre": nombre,
        "precio": precio,
        "cantidad": cantidad
    })

    for producto in inventario:
        print(f"{producto['nombre']} - {producto['precio']}â‚¬ - {producto['cantidad']} unidades")

def calcular_total():
    return sum(p['precio'] * p['cantidad'] for p in inventario)

# Ejemplo uso
