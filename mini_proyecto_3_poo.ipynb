class Tarea:
    def __init__(self, titulo, descripcion):
        self.titulo = titulo
        self.descripcion = descripcion
        self.hecha = False

    def marcar_hecha(self):
        self.hecha = True

    def __str__(self):
        estado = "x" if self.hecha else " "
        return f"[{estado}] {self.titulo} - {self.descripcion}"


class GestorTareas:
    def __init__(self):
        self.tareas = []

    def agregar(self, tarea):
        self.tareas.append(tarea)

    def listar(self):
        for i, t in enumerate(self.tareas, start=1):
            print(i, t)

    def marcar_hecha(self, indice):
        self.tareas[indice - 1].marcar_hecha()


gestor = GestorTareas()
gestor.agregar(Tarea("Comprar", "Ir al s√∫per"))
gestor.agregar(Tarea("Estudiar", "Python POO"))
gestor.agregar(Tarea("Deporte", "Correr 30 min"))

gestor.listar()
gestor.marcar_hecha(2)
print("---")
gestor.listar()
