## üìè PEP 8 (lo m√≠nimo viable)

PEP 8 es la gu√≠a de estilo de Python.

- 4 espacios de indentaci√≥n.

- L√≠neas no demasiado largas (ideal < 88 caracteres).


## ‚úÖ Ejemplo PEP 8 (simple)
