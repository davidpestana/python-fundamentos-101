## üß† ¬øQu√© es una variable?

Una variable es un nombre que referencia un valor en memoria.

En Python no necesitas declarar el tipo.

Ejemplo:

````python
edad = 25
nombre = "Ana"
print(edad)
print(nombre)
````

## üîÑ Tipado din√°mico

Python detecta autom√°ticamente el tipo del valor.

Una variable puede cambiar de tipo:

````python
x = 10
print(type(x))

x = "ahora soy texto"
print(type(x))
````

## üì¶ Tipos primitivos

- `int` ‚Üí n√∫meros enteros
- `float` ‚Üí n√∫meros decimales
- `str` ‚Üí texto
- `bool` ‚Üí verdadero/falso
- `None` ‚Üí ausencia de valor


````python
a = 10          # int
b = 3.14        # float
c = "Python"   # str
d = True        # bool
e = None        # NoneType

print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
````
