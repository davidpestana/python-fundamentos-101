## ‚ûï Operadores aritm√©ticos

- `+` suma
- `-` resta
- `*` multiplicaci√≥n
- `/` divisi√≥n real
- `//` divisi√≥n entera
- `%` m√≥dulo
- `**` potencia


````python
a = 10
b = 3

print("Suma:", a + b)
print("Resta:", a - b)
print("Multiplicaci√≥n:", a * b)
print("Divisi√≥n real:", a / b)
print("Divisi√≥n entera:", a // b)
print("M√≥dulo:", a % b)
print("Potencia:", a ** b)
````

## üîé Operadores de comparaci√≥n

- `==` igual
- `!=` distinto
- `<`, `>`, `<=`, `>=`


````python
x = 5
y = 10

print(x == y)
print(x != y)
print(x < y)
print(x > y)
````

## üß† Operadores l√≥gicos

- `and`
- `or`
- `not`

Devuelven valores booleanos.


````python
a = True
b = False

print(a and b)
print(a or b)
print(not a)
````
