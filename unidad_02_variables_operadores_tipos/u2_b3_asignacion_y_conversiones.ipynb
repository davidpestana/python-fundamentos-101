## üîÅ Operadores de asignaci√≥n compuesta

- `+=`
- `-=`
- `*=`
- `/=`
- `//=`
- `%=`
- `**=`

Son atajos para operar y reasignar.


````python
x = 10
x += 5   # x = x + 5
print(x)

x *= 2   # x = x * 2
print(x)

x -= 4   # x = x - 4
print(x)
````

## üîÑ Conversiones expl√≠citas (casting)

Funciones comunes:

- `int()`
- `float()`
- `str()`
- `bool()`


````python
numero_texto = "10"
numero = int(numero_texto)
print(numero, type(numero))

decimal = float("3.14")
print(decimal, type(decimal))

texto = str(25)
print(texto, type(texto))

print(bool(1))
print(bool(0))
````

## ‚ö†Ô∏è Errores comunes

No se puede mezclar `str` con `int` directamente:


````python
# Esto generar√° error:
# "Edad: " + 20

# Forma correcta:
"Edad: " + str(20)
````

## üîé Casting impl√≠cito

Python puede convertir autom√°ticamente en algunas operaciones:


````python
resultado = 5 + 3.2
print(resultado, type(resultado))
````
