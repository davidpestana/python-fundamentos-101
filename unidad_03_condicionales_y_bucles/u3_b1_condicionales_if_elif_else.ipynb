La condici√≥n debe evaluarse como `True` o `False`.


````python
edad = 20

if edad >= 18:
    print("Eres mayor de edad")
````

## üîÄ if ‚Äì elif ‚Äì else

Permite evaluar m√∫ltiples condiciones.


````python
nota = 7

if nota >= 9:
    print("Sobresaliente")
elif nota >= 7:
    print("Notable")
elif nota >= 5:
    print("Aprobado")
else:
    print("Suspenso")
````

## üîó Condiciones compuestas

Puedes combinar condiciones con `and`, `or`, `not`.


````python
edad = 30
tiene_licencia = True

if edad >= 18 and tiene_licencia:
    print("Puede conducir")
````
