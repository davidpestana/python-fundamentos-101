## üß† Estructura b√°sica del for

```python
for variable in iterable:
    bloque
```

El bucle recorre cada elemento del iterable.


````python
for i in range(5):
    print(i)
````

## üî¢ range()

Formas comunes:

- `range(5)` ‚Üí 0 a 4
- `range(1, 6)` ‚Üí 1 a 5
- `range(1, 10, 2)` ‚Üí paso de 2


````python
for i in range(1, 6):
    print(i)
````

## üîÅ Iterar sobre listas


````python
frutas = ["manzana", "pera", "pl√°tano"]

for fruta in frutas:
    print(fruta)
````

## üî¢ enumerate()

Permite obtener √≠ndice y valor al mismo tiempo.


````python
for indice, fruta in enumerate(frutas):
    print(indice, fruta)
````
