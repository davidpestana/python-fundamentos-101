## ‚õî break

Finaliza el bucle inmediatamente.


````python
for i in range(10):
    if i == 5:
        break
    print(i)
````

## ‚è≠ continue

Salta a la siguiente iteraci√≥n.


````python
for i in range(5):
    if i == 2:
        continue
    print(i)
````

## üîö else en bucles

El `else` se ejecuta si el bucle termina sin `break`.


````python
for i in range(3):
    print(i)
else:
    print("Bucle terminado sin break")
````

## üîÄ match/case (Python 3.10+)

Permite evaluar m√∫ltiples patrones.


````python
dia = "lunes"

match dia:
    case "lunes":
        print("Inicio de semana")
    case "viernes":
        print("Casi fin de semana")
    case _:
        print("D√≠a normal")
````
