## üîé Indexaci√≥n

Los strings son secuencias.

Empiezan en √≠ndice 0.


````python
texto = "Python"

print(texto[0])  # P
print(texto[1])  # y
print(texto[-1]) # n (√∫ltimo)
````

## ‚úÇÔ∏è Slicing

Sintaxis:

```python
texto[inicio:fin]
```

- Incluye inicio
- Excluye fin


````python
texto = "Python"

print(texto[0:3])  # Pyt
print(texto[2:])   # thon
print(texto[:4])   # Pyth
print(texto[-3:])  # hon
````

## üîÑ Paso en slicing

```python
texto[inicio:fin:paso]
```


````python
texto = "Python"

print(texto[::2])   # saltando 1 car√°cter
print(texto[::-1])  # invertir string
````
