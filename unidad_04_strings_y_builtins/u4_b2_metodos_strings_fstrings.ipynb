## üßµ Strings son inmutables

Los m√©todos no modifican el texto original.

Devuelven un nuevo string.


````python
texto = "python"

print(texto.upper())
print(texto)  # original no cambia
````

## üî§ M√©todos comunes

- `upper()`
- `lower()`
- `strip()`
- `replace()`
- `split()`
- `join()`
- `find()`


````python
texto = "  Hola Mundo  "

print(texto.strip())
print(texto.lower())
print(texto.replace("Mundo", "Python"))
print(texto.find("Hola"))
````

## üîÄ split() y join()


````python
frase = "uno,dos,tres"
lista = frase.split(",")
print(lista)

texto_unido = "-".join(lista)
print(texto_unido)
````

## ‚ú® f-strings (formato moderno)

Permiten insertar variables directamente en texto.


````python
nombre = "Ana"
edad = 30

print(f"Hola {nombre}, tienes {edad} a√±os.")
````

Tambi√©n permiten expresiones:


````python
precio = 100
IVA = 0.21

print(f"Total con IVA: {precio * (1 + IVA)}")
````
