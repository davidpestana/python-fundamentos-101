## üì¶ Crear listas

Las listas se definen con corchetes `[]`.


````python
numeros = [1, 2, 3]
nombres = ["Ana", "Luis", "Marta"]

print(numeros)
print(nombres)
````

## üîé Acceso por √≠ndice


````python
print(nombres[0])
print(nombres[-1])
````

## ‚úÇÔ∏è Slicing en listas


````python
print(numeros[0:2])
print(numeros[::-1])
````

## ‚ûï M√©todos principales


````python
lista = [3, 1, 4]

lista.append(5)
print(lista)

lista.insert(1, 9)
print(lista)

lista.remove(4)
print(lista)

ultimo = lista.pop()
print("Extra√≠do:", ultimo)
print(lista)

lista.sort()
print(lista)
````
