## üì¶ Tuplas

Las tuplas son similares a listas pero **inmutables**.


````python
coordenadas = (10, 20)
print(coordenadas)
print(coordenadas[0])
````

## üîÑ Desempaquetado


````python
x, y = coordenadas
print(x, y)
````

## üîó Sets

Colecci√≥n sin duplicados.


````python
numeros = {1, 2, 2, 3, 4}
print(numeros)
````

### Operaciones con sets


````python
a = {1, 2, 3}
b = {3, 4, 5}

print(a.union(b))
print(a.intersection(b))
print(a.difference(b))
````

## üìö Diccionarios

Estructura clave-valor.


````python
persona = {
    "nombre": "Ana",
    "edad": 30
}

print(persona["nombre"])
print(persona.get("edad"))
````

### Iterar sobre diccionarios


````python
for clave, valor in persona.items():
    print(clave, valor)
````
