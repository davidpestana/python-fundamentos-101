## üîÅ Iterar listas


````python
numeros = [1, 2, 3, 4]

for numero in numeros:
    print(numero)
````

## üîÅ Iterar diccionarios


````python
persona = {"nombre": "Ana", "edad": 30}

for clave in persona:
    print(clave, persona[clave])

for clave, valor in persona.items():
    print(clave, valor)
````

## üîÑ Transformar listas con for

Ejemplo: obtener cuadrados.


````python
numeros = [1, 2, 3, 4]
cuadrados = []

for n in numeros:
    cuadrados.append(n ** 2)

print(cuadrados)
````

## ‚ú® List Comprehension

Sintaxis:

```python
[expresi√≥n for elemento in iterable]
```


````python
numeros = [1, 2, 3, 4]
cuadrados = [n ** 2 for n in numeros]
print(cuadrados)
````

## ‚ûï Con condici√≥n


````python
pares = [n for n in range(10) if n % 2 == 0]
print(pares)
````
