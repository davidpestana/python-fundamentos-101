## üîß Par√°metros por defecto


````python
def saludar(nombre="Invitado"):
    return f"Hola {nombre}"

print(saludar())
print(saludar("Ana"))
````

## üè∑ Argumentos por nombre


````python
def presentar(nombre, edad):
    return f"{nombre} tiene {edad} a√±os"

print(presentar(edad=30, nombre="Luis"))
````

## üì¶ *args

Permite recibir m√∫ltiples argumentos.


````python
def sumar_todos(*numeros):
    total = 0
    for n in numeros:
        total += n
    return total

print(sumar_todos(1, 2, 3, 4))
````

## üì¶ **kwargs

Recibe argumentos clave-valor.


````python
def mostrar_info(**datos):
    for clave, valor in datos.items():
        print(clave, valor)

mostrar_info(nombre="Ana", edad=25)
````

## üåç Scope

Las variables dentro de una funci√≥n son locales.


````python
x = 10

def mostrar():
    x = 5
    print(x)

mostrar()
print(x)
````

## üßæ Type hints


````python
def multiplicar(a: int, b: int) -> int:
    return a * b

print(multiplicar(3, 4))
````

## üìò Docstrings


````python
def dividir(a: float, b: float) -> float:
    """Divide dos n√∫meros y devuelve el resultado."""
    return a / b

print(dividir(10, 2))
````
