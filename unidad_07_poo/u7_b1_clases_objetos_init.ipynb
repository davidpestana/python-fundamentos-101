## üîß Constructor __init__

Se ejecuta al crear el objeto.


````python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

p = Persona("Ana", 30)
print(p.nombre)
print(p.edad)
````

## üß† M√©todos de instancia


````python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def saludar(self):
        return f"Hola, soy {self.nombre}"

p = Persona("Luis", 25)
print(p.saludar())
````

```python
class Rectangulo:
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def area(self):
        return self.base * self.altura

    def perimetro(self):
        return 2 * (self.base + self.altura)

r = Rectangulo(4, 5)
print(r.area())
print(r.perimetro())
```
