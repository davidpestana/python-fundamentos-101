## üîí Encapsulaci√≥n (convenci√≥n)

- `_atributo` ‚Üí protegido (convenci√≥n)
- `__atributo` ‚Üí name mangling


````python
class Cuenta:
    def __init__(self, titular, saldo):
        self.titular = titular
        self._saldo = saldo
        self.__pin = "1234"

    def mostrar_saldo(self):
        return self._saldo

c = Cuenta("Ana", 1000)
print(c.mostrar_saldo())
````

## üßæ @property

Permite controlar acceso como si fuera atributo.


````python
class Persona:
    def __init__(self, edad):
        self._edad = edad

    @property
    def edad(self):
        return self._edad

p = Persona(30)
print(p.edad)
````

## üß¨ Herencia


````python
class Animal:
    def __init__(self, nombre):
        self.nombre = nombre

    def hablar(self):
        return "Sonido"

class Perro(Animal):
    def hablar(self):
        return "Guau"

p = Perro("Rex")
print(p.nombre)
print(p.hablar())
````

## üîÅ super()


````python
class Gato(Animal):
    def __init__(self, nombre, color):
        super().__init__(nombre)
        self.color = color

g = Gato("Misu", "Negro")
print(g.nombre, g.color)
````

## ‚≠ê M√©todos especiales


````python
class Libro:
    def __init__(self, titulo):
        self.titulo = titulo

    def __str__(self):
        return f"Libro: {self.titulo}"

l = Libro("Python 101")
print(l)
````

class Cuenta:
    def __init__(self, titular, saldo):
        self.titular = titular
        self.saldo = saldo

    def __str__(self):
        return f"Cuenta de {self.titular}: {self.saldo}‚Ç¨"

class CuentaAhorro(Cuenta):
    def aplicar_interes(self, tasa):
        self.saldo += self.saldo * tasa

c.aplicar_interes(0.05)
print(c)
