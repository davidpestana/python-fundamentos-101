## üîÑ map()

Aplica funci√≥n a cada elemento.


````python
numeros = [1, 2, 3, 4]
resultado = list(map(lambda x: x * 2, numeros))
print(resultado)
````

## üîé filter()

Filtra elementos.


````python
pares = list(filter(lambda x: x % 2 == 0, numeros))
print(pares)
````

## ‚ûó reduce()

Reduce a un solo valor.


````python
from functools import reduce

suma = reduce(lambda a, b: a + b, numeros)
print(suma)
````

## üí§ Generators

Evaluaci√≥n lazy.


````python
gen = (x * 2 for x in range(5))
print(gen)
print(list(gen))
````
