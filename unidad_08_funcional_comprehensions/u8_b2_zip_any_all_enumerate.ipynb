## ðŸ”— zip()

Une elementos de varios iterables en pares (o tuplas).


````python
nombres = ["Ana", "Luis", "Marta"]
edades = [30, 25, 40]

pares = list(zip(nombres, edades))
print(pares)
````

## ðŸ§¾ Convertir zip a diccionario


````python
agenda = dict(zip(nombres, edades))
print(agenda)
````

## âœ… any()

Devuelve True si **algÃºn** elemento es True.


````python
numeros = [1, 3, 5, 8]
hay_par = any(n % 2 == 0 for n in numeros)
print(hay_par)
````

## âœ… all()

Devuelve True si **todos** los elementos son True.


````python
todos_positivos = all(n > 0 for n in numeros)
print(todos_positivos)
````

## ðŸ”¢ enumerate()

Itera devolviendo (Ã­ndice, valor).


````python
frutas = ["manzana", "pera", "uva"]

for i, fruta in enumerate(frutas, start=1):
    print(i, fruta)
````
