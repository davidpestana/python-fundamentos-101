## üìÇ Abrir archivo

Modos comunes:

- "r" ‚Üí lectura
- "w" ‚Üí escritura (sobrescribe)
- "a" ‚Üí a√±adir


````python
with open("ejemplo.txt", "w") as f:
    f.write("Hola mundo\n")
    f.write("Segunda l√≠nea")
````

## üìñ Leer archivo


````python
with open("ejemplo.txt", "r") as f:
    contenido = f.read()

print(contenido)
````

## üìú Leer l√≠nea a l√≠nea


````python
with open("ejemplo.txt", "r") as f:
    for linea in f:
        print(linea.strip())
````

## ‚ö†Ô∏è Manejo de excepciones
