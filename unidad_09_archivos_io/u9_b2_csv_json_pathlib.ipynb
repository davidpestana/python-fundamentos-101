## üìä CSV con m√≥dulo csv


````python
import csv

datos = [
    ["Nombre", "Edad"],
    ["Ana", 30],
    ["Luis", 25]
]

with open("personas.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(datos)
````

````python
with open("personas.csv", "r") as f:
    reader = csv.reader(f)
    for fila in reader:
        print(fila)
````

## üóÇ JSON

Ideal para estructuras clave-valor.


````python
import json

persona = {
    "nombre": "Ana",
    "edad": 30
}

with open("persona.json", "w") as f:
    json.dump(persona, f)
````

````python
with open("persona.json", "r") as f:
    datos = json.load(f)

print(datos)
print(datos["nombre"])
````

## üìÅ pathlib

Manejo moderno de rutas.


````python
from pathlib import Path

ruta = Path("datos")
ruta.mkdir(exist_ok=True)

archivo = ruta / "archivo.txt"
archivo.write_text("Contenido con pathlib")

print(archivo.read_text())
````

import csv
import json

productos = [
    {"nombre": "Laptop", "precio": 1000},
    {"nombre": "Mouse", "precio": 25}
]

# CSV
with open("productos.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["nombre", "precio"])
    for p in productos:
        writer.writerow([p["nombre"], p["precio"]])

# JSON
with open("productos.json", "w") as f:
    json.dump(productos, f)
