## üß† Concurrencia vs Paralelismo

- Concurrencia: tareas se alternan.
- Paralelismo: tareas se ejecutan simult√°neamente.

Python tiene GIL ‚Üí threading no siempre escala en CPU.
Async es ideal para I/O (esperas, red, archivos).


## ‚ö° Coroutine b√°sica


````python
import asyncio

async def saludar():
    print("Hola")

asyncio.run(saludar())
````

## ‚è≥ await con sleep

Simulamos operaci√≥n I/O.


````python
async def tarea():
    print("Inicio")
    await asyncio.sleep(1)
    print("Fin")

asyncio.run(tarea())
````

## üîÑ Ejecutar m√∫ltiples tareas


````python
async def tarea(nombre, tiempo):
    print(f"Inicio {nombre}")
    await asyncio.sleep(tiempo)
    print(f"Fin {nombre}")

async def main():
    await asyncio.gather(
        tarea("A", 2),
        tarea("B", 1)
    )

asyncio.run(main())
````
