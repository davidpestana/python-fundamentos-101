## üöÄ create_task()

Permite lanzar tareas que se ejecutan en segundo plano.


````python
import asyncio

async def tarea(nombre):
    print(f"Inicio {nombre}")
    await asyncio.sleep(1)
    print(f"Fin {nombre}")

async def main():
    t1 = asyncio.create_task(tarea("A"))
    t2 = asyncio.create_task(tarea("B"))
    await t1
    await t2

asyncio.run(main())
````

## ‚è± Timeout

Cancelar tarea si tarda demasiado.


````python
async def larga():
    await asyncio.sleep(5)
    return "Terminado"

async def main():
    try:
        resultado = await asyncio.wait_for(larga(), timeout=2)
        print(resultado)
    except asyncio.TimeoutError:
        print("Timeout!")

asyncio.run(main())
````

## üö¶ Semaphore

Limita n√∫mero de tareas simult√°neas.


````python
sem = asyncio.Semaphore(2)

async def trabajo(nombre):
    async with sem:
        print(f"Inicio {nombre}")
        await asyncio.sleep(1)
        print(f"Fin {nombre}")

async def main():
    await asyncio.gather(*(trabajo(f"T{i}") for i in range(5)))

asyncio.run(main())
````

import asyncio

sem = asyncio.Semaphore(2)

async def trabajo(nombre, tiempo):
    async with sem:
        try:
            await asyncio.wait_for(asyncio.sleep(tiempo), timeout=4)
            print(f"{nombre} completado")
        except asyncio.TimeoutError:
            print(f"{nombre} cancelado por timeout")

async def main():
    await asyncio.gather(
        trabajo("A", 1),
        trabajo("B", 5),
        trabajo("C", 2),
        trabajo("D", 3)
    )

asyncio.run(main())
