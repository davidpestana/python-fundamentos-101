```python
from datetime import datetime, timedelta
from collections import Counter
import itertools

colores = ["Rojo", "Azul"]
tallas = ["S", "M"]

hoy = datetime.now()

resultado = []
for color, talla in itertools.product(colores, tallas):
    resultado.append({
        "producto": color,
        "talla": talla,
        "fecha": hoy + timedelta(days=7)
    })

conteo = Counter([r["talla"] for r in resultado])

print(resultado)
print(conteo)
```
