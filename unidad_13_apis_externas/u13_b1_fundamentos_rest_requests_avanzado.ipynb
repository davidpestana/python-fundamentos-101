## üîé GET con query params


````python
import requests

params = {"userId": 1}
response = requests.get("https://jsonplaceholder.typicode.com/posts", params=params)

print(response.url)
print(response.status_code)
print(response.json()[:2])
````

## ‚ûï POST con JSON


````python
nuevo_post = {
    "title": "Nuevo",
    "body": "Contenido",
    "userId": 1
}

response = requests.post(
    "https://jsonplaceholder.typicode.com/posts",
    json=nuevo_post
)

print(response.status_code)
print(response.json())
````

## üîÅ PUT (simulado)


````python
actualizacion = {"title": "Actualizado"}

response = requests.put(
    "https://jsonplaceholder.typicode.com/posts/1",
    json=actualizacion
)

print(response.status_code)
print(response.json())
````

## ‚ùå DELETE


````python
response = requests.delete("https://jsonplaceholder.typicode.com/posts/1")
print(response.status_code)
````

## ‚ö†Ô∏è Manejo b√°sico de errores


````python
response = requests.get("https://jsonplaceholder.typicode.com/invalid")

if response.status_code == 200:
    print("OK")
else:
    print("Error:", response.status_code)
````

def get_posts_por_usuario(user_id):
    response = requests.get(
        "https://jsonplaceholder.typicode.com/posts",
        params={"userId": user_id}
    )

    if response.status_code != 200:
        raise Exception("Error en API")

    posts = response.json()
    return [p["title"] for p in posts]

print(get_posts_por_usuario(1))
